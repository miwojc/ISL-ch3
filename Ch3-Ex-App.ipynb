## 9. Multiple linear regression on the Auto data set

Scatterplot matrix which includes all of the variables in the data set

```r
pairs(Auto)
```

Matrix of correlations between variabes using cor() function, exclude qualitative variable: name

```r
AutoNames <- names(Auto) %in% c("name") 
AutoLessNames <- Auto[!AutoNames]
```

```r
cor(AutoLessNames)
```

Multiple linear regression using lm() function on all predictors except name with summary() functions for printing results

```r
summary(lm(mpg~.,data=AutoLessNames))
```

There is relationship between mpg and predictors: R=0.8215, F=254.4, p-value is less than 0.05.<br>
Predictors with statistically significant relationship: displacement, weight, year, origin.<br>
Coefficient for the year variable suggests that there is positive relationship between mpg and year.

Diagnostic plots:

```r
par(mfrow=c(2,2))
plot(lm(mpg~.,data=AutoLessNames))
```

Plots suggest non-linear data. Data suggest outliers on Scale-Location and high leverage points at Residuals vs Leverage plot.

Fit linear regression model with interaction plots:

```r
summary(lm(mpg~.+cylinders:displacement+cylinders:horsepower+cylinders:weight+cylinders:acceleration+cylinders:year+cylinders:origin+displacement:horsepower+displacement:weight+displacement:acceleration+displacement:year+displacement:origin+horsepower:weight+horsepower:acceleration+horsepower:year+horsepower:origin+weight:acceleration+weight:year+weight:origin+acceleration:year+acceleration:origin+year:origin, data=AutoLessNames))
```

Interactions that appear to be statistically significant at 0.05 level: displacement:year, acceleration:year and acceleration: origin. 

Various transformation of the variables

```r
summary(lm(mpg~acceleration))
summary(lm(mpg~acceleration+I(acceleration^2)))
summary(lm(mpg~acceleration+I(sqrt(acceleration))))
summary(lm(mpg~acceleration+I(log(acceleration))))
```

None of the transformation (x2, log(x) or sqrt(x)) significantly improved model for acceleration based on adjusted R-squared

## 10. Carseats data set

```r
names(Carseats)
```

```r
summary(Carseats)
```

Multiple regression model to predict Sales using Price, Urban and US.

```r
attach(Carseats)
lm.fit1 = lm(Sales~Price+Urban+US)
```

```r
summary(lm.fit1)
```

Coefficients:<br>
Price: negative relationsiop to Sales, statistically significant as p is low<br>
Urban: Urban locations contribute to lower Sales, statistically not significant as p is high<br>
US: US locations contribute to higher Sales, statistically significant as p is low

Model: Sales = 13.043469 - 0.054459*Price - 0.021916*UrbanYes + 1.200573*USYes

Urban predictor has high p-value and it's not statistically significant

Model only with predictors for which there is evidence of association with the outcome:

```r
lm.fit2 = lm(Sales~Price+US)
summary(lm.fit2)
```

Model with Price and US predictors only has higher F-statistic, Adjusted R-squared is a bit better, p-value same. So getting rid of one predictor makes the model simpler, which is good.

95% confidence intervals for coefficients using confint() function

```r
confint(lm.fit2)
```

Diagnostic plots:

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

From diagnostic plots there is evidence of outliers and high leverage points

## 11 t-statistic for the null hyphotesis

Generate a preditor x and response y as follows:

```r
set.seed(1)
x=rnorm(100)
y=2*x+rnorm(100)
```

Simple linear regression y onto x without intercept

```r
lm.fit111 = lm(y~x+0)
```

```r
summary(lm.fit111)
```

Coeficient estimate is 1.9939, standard error = 0.1065, t-statistic = 18.73, p-value < 2e-16<br>
The relationship x and y is positive and statistically significant.

Simple linear regression x onto y without intercept

```r
lm.fit112 = lm(x~y+0)
summary(lm.fit112)
```

Coeficient estimate is 0.39111, standard error = 0.02089, t-statistic = 18.73, p-value < 2e-16<br>
The relationship y and x is positive and statistically significant.

For both analyses the t value, p-value, Adjusted R-squared and F-statistic are the same. there is a difference in coefficient and standard error.

Regression with an intercept: t-statistic for regression of y onto x and x onto y

```r
lm.fit113 = lm(y~x)
summary(lm.fit113)
```

```r
lm.fit114 = lm(x~y)
summary(lm.fit114)
```

For both regressions the t-statistic, for both x and y is 18.56
