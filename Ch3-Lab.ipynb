## Multiple Linear Regression

Multiple linear regression with lsat and age as predictors

```r
lm.fit=lm(medv~lstat+age,data=Boston)
summary(lm.fit)
```

Multiple linear regression with all 13 predictors from Boston data set

## Non-linear Transformations of the Predictors

X*X = I(X^2)

```r
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

Use anova() funtion to further quantify the extent to which the quadratic fit is superior to the linear fit.

```r
lm.fit=lm(medv~lstat)
anova(lm.fit,lm.fit2)
```

Model with interaction performs better as expected, since non-linear relationship between medv and lstat.

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

Little discernible pattern in the residuals when lstat^2 is used

Usinig poly() funciton to create 5th level polylnomial fit

```r
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```

We can use log transformation

```r
summary(lm(medv~log(rm),data=Boston))
```
